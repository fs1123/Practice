```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures
import seaborn as sns
```

```python
df_all = pd.read_csv("LoanStats_2016Q1_small.csv")
```

```python
df_part = df_all[['loan_amnt', 'annual_inc']]
```

```python
y = df_part['annual_inc'].values.reshape(-1,1)
X = df_part['loan_amnt'].values.reshape(-1,1)
```

## Plot with degree = 1

```python
# Degree = 1, fit model
regr = linear_model.LinearRegression()
regr.fit(X,y);
```

Calculate $Y = \hat{\beta_{0}} + \hat{\beta_{1}}X$ with `regr.predict()`

```python
# Calculate 
X_grid = np.linspace(X.min(), X.max(), 10**2).reshape(-1,1)
pred1 = regr.predict(X_grid)
```

```python
# Plot
fig, ax = plt.subplots(1,1,figsize=(8,6))
ax.scatter(X,y)
ax.plot(X_grid, pred1, lw=3, c='k', linestyle='--', label='degree 1')
ax.legend();
```

## Plot with degree = 2

```python
# PolynomialFeatures
X_poly = PolynomialFeatures(2).fit_transform(X)
```

```python
regr = linear_model.LinearRegression()
regr.fit(X_poly, y);
```

Calculate $Y = \hat{\beta_{0}} + \hat{\beta_{1}}X + \hat{\beta_{2}}X^{2}$ with `regr.predict()`

```python
X_grid = np.linspace(X.min(), X.max(), 10**2).reshape(-1,1)
X_grid_poly = PolynomialFeatures(2).fit_transform(X_grid)
pred2 = regr.predict(X_grid_poly)
```

```python
# Plot
fig, ax = plt.subplots(1,1,figsize=(8,6))
ax.scatter(X,y, alpha=0.3)
ax.plot(X_grid, pred2, lw=3, c='r', linestyle="-.", label='degree2')
ax.legend();
```

## Plot Degree 1 & Degree 2

```python
fig, ax = plt.subplots(1,1,figsize=(8,6))
ax.scatter(X,y, alpha=0.3)
ax.plot(X_grid, pred1, lw=3, c='k', linestyle='--', label='degree 1')
ax.plot(X_grid, pred2, lw=3, c='r', linestyle='-.', label='degree 2');
ax.legend();
```

## Plotting with `sns.regplot`

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12,6))
# Degree 1
sns.regplot(X.ravel(),y.ravel(), fit_reg=True, ax=ax1, scatter_kws={'alpha':0.3}, line_kws={'color':'k'})

# Degree 2
sns.regplot(X.ravel(),y.ravel(), fit_reg=True, order=2, ax=ax2, scatter_kws={'alpha':0.3}, line_kws={'color':'r'});
```

### Plot at the same time

```python
fig, ax = plt.subplots(1,1, figsize=(8,6))
# Degree 1
sns.regplot(X.ravel(),y.ravel(), fit_reg=True, ci=None, ax=ax, scatter_kws={'alpha':0.3}, line_kws={'color':'k'})

# Degree 2
sns.regplot(X.ravel(),y.ravel(), scatter=False, fit_reg=True, ci=None, order=2, ax=ax, line_kws={'color':'r'});
```
